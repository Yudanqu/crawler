```python
import requests
import re

def getHtmlText(url):
    print(url)
    try:
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.12 Safari/537.36"}
        r = requests.get(url, headers=headers, timeout=500)
        print(r.status_code)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        # print(r.text)
        return r.text
    except:
        print("爬取失败")
        return ""
    
def parsePage(lis, html):
    try:
        plt = re.findall("view_price", html) # "view_price":"138.00"
        tlt = re.findall(r'\"raw_title\"\:\".*?\"', html)
        print(plt, tlt)
        for i in range(len(plt)):
            price = eval(plt[i].split(':')[1])  # eval函数可以使最外层的引号去掉
            title = eval(tlt[i].split(':')[1])
            lis.append([price, title])
            print([price, title])
    except:
        print("")
    
def printGoodsList(lis):
    tplt = "{:4}\t{:8}\t{:16}"
    print(tplt.format("序号", "价格", "商品名称"))
    count = 0
    for g in lis:
        count = count + 1
        print(tplt.format(count, g[0], g[1]))
    
if __name__ == "__main__":
    goods = "书包"
    depth = 2
    start_url = "https://s.taobao.com/search?q=" + goods
    infoList = []
    for i in range(depth):
        try:
            url = start_url + "&s=" + str(44*i)  # 对于不同页面内的商品的爬取
            # print(url)
            html = getHtmlText(url)
            # print(html[:500])
            parsePage(infoList, html)
        except:
            continue
    printGoodsList(infoList)
```
