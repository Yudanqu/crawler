```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
for i in range(1, 101):
    try:
        response = requests.get("https://xkcd.com/{}/".format(i))
        html = response.text
        response.raise_for_status()
        soup = BeautifulSoup(html, "html.parser")
        _title = soup.title.contents[0][6:]
        print(_title)
        obj = soup.find('div', id="comic")
        print(str(obj.contents[1]))
        title = re.search('title="(.*?)"', str(obj.contents[1]))[1]
        url = re.search('src="(.*?)"', str(obj.contents[1]))[1]
        url = f"https:{url}"
        pic = requests.get(url)
        name = 'pics\\' + title + '.jpg'
        with open(name, 'wb') as f:
            f.write(pic.content)
    except:
        print("Fail")
```
