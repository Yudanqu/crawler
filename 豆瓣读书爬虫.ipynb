```python
import requests
import re
import pandas as pd
```

```python
url = "https://book.douban.com/"
def getHtmlText(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'
    }
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        response.encoding = response.apparent_encoding
        return response.text
    except:
        print("Fail")
        return
    
html = getHtmlText(url)
print(html[:500])
```

```python
re_books = re.compile('<ul class="list-col list-col5 list-express slide-item">(.*?)</ul>', re.S)
content = re_books.findall(html)
print(content[0])
```

```python
data = pd.DataFrame(columns=['title', 'author', 'abstract', 'href', 'publisher'])

re_book = re.compile('<li class="">(.*?)</li>', re.S)
bookList = re_book.findall(content[0])
for book in bookList:
    count = 0
    count += 1
    href = re.search('href="(.*?)"', book)
    href = href.group(1)
    title = re.search('<h4 class="title">(.*?)</h4>', book, re.S)
    title = title.group(1).split()[0]
    author = re.search('<span class="author">(.*?)</span>', book, re.S)
    author = ' '.join(author.group(1).split())
    publisher = re.search('<span class="publisher">(.*?)</span>', book, re.S)
    publisher = ' '.join(publisher.group(1).split())
    abstract = re.search('<p class="abstract">(.*?)</p>', book, re.S)
    abstract = ' '.join(abstract.group(1).split())
    abstract = re.sub('【内容简介】', '', abstract)
    new = pd.DataFrame({"title":title, "author":author, "abstract":abstract, "href":href, "publisher":publisher}, index=["0"])
    data = data.append(new, ignore_index=True)
data
```
